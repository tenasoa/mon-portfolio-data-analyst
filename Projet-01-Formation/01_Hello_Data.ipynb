## Création de données simple

```python
# Créer un dataset simple
data = {
    'Mois' : ['Janvier', 'Février', 'Mars', 'Avril', 'Mais', 'Juin'],
    'Ventes' : [1200, 1500, 1100, 1800, 2000, 1900],
    'Visiteurs' : [5000, 6200, 4800, 7500, 8200, 8000]
}

df = pd.DataFrame(data)
print("\n Mes premières données : ")
print(df)

# Statistique
print("\n Statistiques descriptives : ")
print(df.describe())
```

## Visualisation

```python
# Créer un graphique
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Graphique 1 : Ventes
axes[0].plot(df['Mois'], df['Ventes'], marker='o', linewidth=2, color='#2E86AB')
axes[0].set_title('Évolution des ventes (€)', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Ventes (€)')
axes[0].grid(alpha=0.3)

# Graphique 2 : Visiteurs vs Ventes
axes[1].bar(df['Mois'], df['Visiteurs'], alpha=0.7, label='Visiteurs', color='#A23B72')
axes2 = axes[1].twinx()
axes2.plot(df['Mois'], df['Ventes'], color='#F18F01', marker='o', linewidth=2, label='Ventes')
axes[1].set_title('Visiteurs vs Ventes', fontsize=14, fontweight='bold')
axes[1].set_ylabel('Nombre de visiteurs')
axes2.set_ylabel('Ventes (€)', color='#F18F01')

plt.tight_layout()
plt.show()

print("✓ Graphiques générés avec succès !")
```

## Analyse simple

```python
# Calculs
total_ventes = df['Ventes'].sum()
vente_moyenne = df['Ventes'].mean()
meilleur_mois = df.loc[df['Ventes'].idxmax()]

print("\n🔍 Analyses rapides :")
print(f"Total des ventes : {total_ventes}€")
print(f"Vente moyenne : {vente_moyenne:.0f}€")
print(f"Meilleur mois : {meilleur_mois['Mois']} avec {meilleur_mois['Ventes']}€")

# Insights
correlation = df['Visiteurs'].corr(df['Ventes'])
print(f"\nCorrélation Visiteurs/Ventes : {correlation:.2f}")
if correlation > 0.8:
    print("→ Insight : Plus il y a de visiteurs, plus on vend ! 📈")
```
